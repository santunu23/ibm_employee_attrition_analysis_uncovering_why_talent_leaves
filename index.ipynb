# Overview
This project analyzes the IBM HR Analytics dataset to identify the core drivers of employee attrition. By leveraging SQL for complex data querying and Google BigQuery for high-scale data processing and visualization, the study provides data-driven insights into why talent leaves and the subsequent financial impact on the organization.

# 1. Ask (Define the Problem)
The primary goal of this phase is to define the business problem and the objectives of the analysis.

* **Business Problem:**

    Why are high-performing and skilled employees leaving the organization (Employee Attrition)?

* **Key Questions to Address:**

    1. Is low compensation a primary driver for employees quitting their jobs?
     
    2. How significant is the impact of overtime and excessive workload on attrition?
     
    3. Which specific departments or age groups are at the highest risk of leaving?
     
    4. Are new hires more likely to leave compared to long-term employees (Tenure Analysis)?

* **Objectives:**

    * To provide actionable and strategic recommendations for improving Employee Retention.
    
    * To calculate and visualize the **Estimated Financial Loss** incurred by the organization due to employee turnover.

# 2. Prepare (Data Preparation)
In this phase, we discuss the data source, its structure, and the environment used for analysis.

* **Data Source**: The dataset used for this project is the IBM HR Analytics Employee Attrition & Performance dataset, sourced from Kaggle.

* **Data Storage & Environment**: To handle large-scale data efficiently and demonstrate SQL capabilities, the dataset was uploaded to and managed within Google BigQuery.

* **Data Structure & Relevance**: The dataset contains critical variables such as Monthly Income, Age, Overtime status, Department, and Tenure (Years in Current Role). These features are essential for answering our core research questions and identifying the drivers of attrition.

# **3. Process (Data Processing)**
This phase focuses on cleaning the data and ensuring it is in the correct format for analysis.

* **SQL Integration**: High-performance data querying was performed using Google BigQuery.

* **Data Cleaning**: We performed checks for null values and verified consistency across columns. Data types were ensured for accuracy, such as   converting the Attrition column into a Boolean format for easier calculation.

* **Tools Used**: Google BigQuery (SQL) for data manipulation, Kaggle Notebook (Python) for documentation, and SQL as the primary engine for analysis.

SELECT Gender, count(*) AS Total_number FROM `mybigquerypractice-446408.hr_analytics.hr_employee`

SELECT 
  CASE
    WHEN Age<30 THEN 'Under 30'
    WHEN Age BETWEEN 30 AND 45 THEN '30-45'
    ELSE 'Over 45'
  END AS Age_Group,
  Attrition,
  COUNT(*) AS Employee_Count, 
  ROUND(AVG(MonthlyIncome),2) AS Avg_Monthly_Income
  FROM `mybigquerypractice-446408.hr_analytics.hr_employee`
  GROUP BY Age_Group, Attrition
  ORDER BY Age_Group, Attrition;

  SELECT 
    Department,
    COUNT(*) AS Total_Employees,
    SUM(CASE WHEN Attrition = TRUE THEN 1 ELSE 0 END) AS Left_count,
    ROUND(AVG(CASE WHEN Attrition = TRUE THEN 1 ELSE 0 END) * 100, 2) as Attrition_Rate
  FROM `mybigquerypractice-446408.hr_analytics.hr_employee`
  GROUP BY Department
  ORDER BY Attrition_Rate DESC;

  -- Findout Promotion
  SELECT
   YearsInCurrentRole,
   COUNT(*) AS Total_Employee,
   SUM(CASE WHEN Attrition = True THEN 1 ELSE 0 END) AS Left_Count,
   ROUND(AVG(CASE WHEN Attrition=TRUE THEN 1 ELSE 0 END) * 100,2) AS Attrition_Rate
  FROM `mybigquerypractice-446408.hr_analytics.hr_employee` 
  group by YearsInCurrentRole
  ORDER BY YearsInCurrentRole;

  -- Findout loses during Attrition
    SELECT 
      Department,
      COUNT(*) AS Employees_Left,
      -- Findout Yearly salary to multiply Monthly salary with 12 and then findout lose to multiply with 1.5
      ROUND(SUM(MonthlyIncome * 12 * 1.5), 2) AS Estimated_Financial_Loss_USD
  FROM `mybigquerypractice-446408.hr_analytics.hr_employee`
  WHERE Attrition = true
  GROUP BY Department
  ORDER BY Estimated_Financial_Loss_USD DESC;
